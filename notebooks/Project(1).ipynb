```python

# plt.xticks(x='city', labels, rotation=0);
reshaped_labels = [arabic_reshaper.reshape(city) for city in aqar['city']]
# Apply bidi algorithm for right-to-left display
bidi_labels = [get_display(label) for label in reshaped_labels]
sns.countplot(aqar, x='city', hue='price', palette='dark:green', legend = False)
# Plot
plt.xticks(aqar['city'], bidi_labels, rotation=0)  # Proper Arabic labels
plt.xlabel("City")
ax = plt.gca()
ax.set_xticklabels([label.get_text()[::-1] for label in ax.get_xticklabels()]);
```
